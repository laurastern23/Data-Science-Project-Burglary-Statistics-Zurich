Anzahl Einbr√ºche im Jahr 2023

```python
import matplotlib.pyplot as plt
import pandas as pd

# CSV-Datei laden
data = pd.read_csv(r'C:\Users\laura\OneDrive\10_Master\DI_Data Science Fundamentals\Git\Data-Science-Project-Burglary-Statistics-Zurich\Cleaned_CSV_files\Einbrueche_Cleaned.csv')

filtered_data = data[(data['Tatbestand'] == 'Einbruchdiebstahl') & (data['Ausgangsjahr'] == 2023)]

# Die Top 10 Gemeinden mit den meisten Einbr√ºchen
top_gemeinden = filtered_data.groupby('Gemeindename')['Straftaten_total'].sum().nlargest(10).reset_index()

# Diagramm erstellen
plt.figure(figsize=(20, 6))
plt.bar(top_gemeinden['Gemeindename'], top_gemeinden['Straftaten_total'])
plt.xlabel('Gemeindename')
plt.ylabel('Anzahl von Einbr√ºchen')
plt.title('Anzahl von Einbr√ºchen pro Gemeinde')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Einkommensverteilung der Gemeinden

```python
import matplotlib.pyplot as plt
import pandas as pd

# CSV-Datei laden
data = pd.read_csv(r'C:\Users\laura\OneDrive\10_Master\DI_Data Science Fundamentals\Git\Data-Science-Project-Burglary-Statistics-Zurich\Cleaned_CSV_files\Einkommensdaten_KantonZurich_2009_2023_Cleaned.csv')

filtered_data = data[data['INDIKATOR_JAHR'] > 2014]

einkommen_pro_gemeinde = filtered_data.groupby('GEBIET_NAME')['INDIKATOR_VALUE'].sum().nlargest(20).reset_index()

# X- und Y-Werte aus der CSV-Datei extrahieren
x = filtered_data['GEBIET_NAME']
y = filtered_data['INDIKATOR_VALUE']

# Diagramm erstellen
plt.figure(figsize=(20, 6))
plt.bar(x, y)
plt.xlabel('Gemeindename')
plt.ylabel('Einkommen in CHF')
plt.title('Einkommen pro Gemeinde')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
