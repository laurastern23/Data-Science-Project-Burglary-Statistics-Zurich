```python
import requests    
import json         
import pandas as pd 
import os

package = 'bevolkerungsdichte-einw-km2'

# Base url for the open data swiss API
base_url = 'https://opendata.swiss/api/3/action/package_show?id='

# Construct the url including package 
package_information_url = base_url + package

# HTTP request
package_information = requests.get(package_information_url)

# Use the json module to load CKAN's response into a dictionary
package_dict = json.loads(package_information.content)

# Check the contents of the response.
assert package_dict['success'] is True  
package_dict = package_dict['result']   
print(package_dict)             
```

```python
# Get the url for the data from the dictionary
data_url = package_dict['resources'][0]['url']
print('Data url:' + data_url)

# Print the data format
data_format = package_dict['resources'][0]['format']
print('Data format:' + data_format)
```

```python
csv = ['comma-separated-values', 'CSV', 'csv']
if any(s in data_format for s in csv):     
     df = pd.read_csv(data_url)
else:
    print('Sorry, the data format is not supported')
df
```

```python
df.drop(['THEMA_NAME','SET_NAME','SUBSET_NAME','INDIKATOR_ID','EINHEIT_KURZ','Unnamed: 11'], axis=1, inplace=True)
df
```

```python
Jahresindizes_zu_loeschen = df[df['INDIKATOR_JAHR'] < 2009].index
df.drop(Jahresindizes_zu_loeschen, inplace=True)
df
```

```python
# Loesche alle Indizes, die den Wert '0' in der Spalte 'BFS_NR' haben da diese zusammengefasste Informationen pro Bezirk/Region enthalten
indizes_zu_loeschen = df[df['BFS_NR'] == 0].index
df.drop(indizes_zu_loeschen, inplace=True)
df
```

```python
#Einige Datensätze beinhalten im Gemeindenamen (bis *) um zu indizieren dass die Rechnung sich auf die angegbenen JAhre bezieht. Um Redundanzen zu vermeiden, werden diese entfernt
indizesBis_zu_loeschen = df[df['GEBIET_NAME'].str.contains("\(bis ", na=False)].index
df.drop(indizesBis_zu_loeschen, inplace=True)
df
```

```python
# Vereinheitlichen der Gemeindenamen
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Aeugst a.A.', 'Aeugst', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Affoltern a.A.', 'Affoltern', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Hausen a.A.', 'Hausen', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Kappel a.A.', 'Kappel', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Wettswil a.A.', 'Wettswil', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Berg a.I.', 'Berg', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Buch a.I.', 'Buch', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Thalheim a.d.Th.', 'Thalheim', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Langnau a.A.', 'Langnau', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Oetwil a.S.', 'Oetwil', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Uetikon a.S.', 'Uetikon', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Ellikon a.d.Th.', 'Ellikon', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Oetwil a.d.L.', 'Oetwil an der Limmat', regex=False)
df['GEBIET_NAME'] = df['GEBIET_NAME'].str.replace('Aesch ZH', 'Aesch', regex=False)
df
```

```python
current_directory = os.getcwd()

file_path = os.path.join(current_directory, "Bevoelkerungsdichte_KantonZurich_2010_2023.csv")
df.to_csv(file_path, index=False)

print(f"The DataFrame has been saved to '{file_path}'.")
```
